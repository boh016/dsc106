```python
chicken = monthly[['CF-NE','CF-SW','CF-NW','CF-SE','CF-C','Month, Year']].set_index('Month, Year')
monthly['CF_AVG'] = list(chicken.apply(np.mean,axis=1))
monthly.CF_AVG.plot()
```

```python
burger = monthly[['HM-NE','HM-SW','HM-NW','HM-SE','HM-C','Month, Year']].set_index("Month, Year")
monthly['HM_AVG'] = list(burger.apply(np.mean,axis=1))
monthly.HM_AVG.plot()
```

```python
fish = monthly[['FF-NE','FF-SW','FF-NW','FF-SE','FF-C','Month, Year']].set_index("Month, Year")
monthly['FF_AVG'] = list(fish.apply(np.mean,axis=1))
monthly.FF_AVG.plot()
```

```python
NE = [i for i in monthly.columns if "-NE" in i]
NW = [i for i in monthly.columns if "-NW" in i]
C = [i for i in monthly.columns if "-C" in i]
SW = [i for i in monthly.columns if "-SW" in i]
SE = [i for i in monthly.columns if "-SE" in i]
```

```python
NE_month = monthly[NE]
NW_month = monthly[NW]
SW_month = monthly[SW]
SE_month = monthly[SE]
C_month = monthly[C]
monthly['NE_AVG'] = list(NE_month.apply(np.mean,axis=1))
monthly['SE_AVG'] = list(SE_month.apply(np.mean,axis=1))
monthly['NW_AVG'] = list(NW_month.apply(np.mean,axis=1))
monthly['SW_AVG'] = list(SW_month.apply(np.mean,axis=1))
monthly['C_AVG'] = list(C_month.apply(np.mean,axis=1))
```

```python
monthly[['Month, Year','HM_AVG','CF_AVG','FF_AVG','NE_AVG','NW_AVG','C_AVG','SW_AVG','SE_AVG']].to_csv("monthly_avg.csv",index=False)
```

```python
daily.Day = pd.to_datetime(daily.Day)
daily_hm = daily[['HM-NE','HM-SW','HM-NW','HM-SE','HM-C','Unnamed: 0','Day']]
daily_cf = daily[['CF-NE','CF-SW','CF-NW','CF-SE','CF-C','Unnamed: 0','Day']]
daily_ff = daily[['FF-NE','FF-SW','FF-NW','FF-SE','FF-C','Unnamed: 0','Day']]

daily_hm
```

```python
hm = daily[['HM-NE','HM-SW','HM-NW','HM-SE','HM-C']].apply(np.mean,axis=1)
cf = daily[['CF-NE','CF-SW','CF-NW','CF-SE','CF-C']].apply(np.mean,axis=1)
ff = daily[['FF-NE','FF-SW','FF-NW','FF-SE','FF-C']].apply(np.mean,axis=1)
daily['HM_AVG'] = list(hm)
daily['CF_AVG'] = list(cf)
daily['FF_AVG'] = list(ff)
daily['NE_AVG'] = list(daily[NE].apply(np.mean,axis=1))
daily['SE_AVG'] = list(daily[SE].apply(np.mean,axis=1))
daily['NW_AVG'] = list(daily[NW].apply(np.mean,axis=1))
daily['SW_AVG'] = list(daily[SW].apply(np.mean,axis=1))
daily['C_AVG'] = list(daily[C].apply(np.mean,axis=1))
daily[['HM_AVG','CF_AVG','FF_AVG','NE_AVG','SE_AVG','NW_AVG','SW_AVG','C_AVG','Day']].to_csv('daily_avg.csv',index=False)
```
